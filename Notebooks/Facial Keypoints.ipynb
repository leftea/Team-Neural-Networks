###Preprocessing:

Kaggle links to a great tutorial - Daniel Nouri, the creator of nolearn (A python package that wraps neural network libraries) did a tutorial on using Convolutional Neural Networks to predict facial keypoints (on this dataset). In his tutorial he has sample code for loading and preprocessing the data, as well as some classes that can be used to implement Early Stopping (more on that in a bit...) as well as adjusting variables in the network as it trains (more on that in a bit...). I will use his [load() function from the tutorial](http://danielnouri.org/notes/2014/12/17/using-convolutional-neural-nets-to-detect-facial-keypoints-tutorial/#the-data)

The following is done to process the data and get it ready to be modeled:

    1. Data is loaded into a pandas dataframe to be processed.
    2. The image column is converted into a numpy array. Originally it was string sperated pixel values, but we want a numpy array of floats so that we can manipulate it and use it for our neural network models.
    3. If Specific columns are specified, only the columns specified are loaded from the data for processing. This comes in handy later when I train multiple networks for specific keypoints!
    4. If drop missing is true, Columns with missing values are dropped (Leaving us with a little over ~2200 samples vs. the original ~7000
    5. The pixel values scaled from [0, 255] to [0,1]
    6. If not loading test data, coordinate output values are scaled from [0,96] to [-1,1]
    7. The Data is shuffled using a random seed for consistency during testing.
    8. Numpy arrays for X (Image inputs) and y (target coordinate outputs) are returned, or just the X if it loading the kaggle test data.


Here is the code that is used:

```python
#Data files
FTRAIN = '../Data/training.csv'
FTEST = '../Data/test.csv'

        
def load(test=False, cols=None, drop_missing=True):
    """Loads data from FTEST if *test* is True, otherwise from FTRAIN.
    Pass a list of *cols* if you're only interested in a subset of the
    target columns. By default, samples with missing coordinates are dropped.
    """
    fname = FTEST if test else FTRAIN
    df = read_csv(os.path.expanduser(fname))  # load pandas dataframe

    # The Image column has pixel values separated by space; convert
    # the values to numpy arrays:
    df['Image'] = df['Image'].apply(lambda im: np.fromstring(im, sep=' '))

    if cols:  # get a subset of columns
        df = df[list(cols) + ['Image']]
   
    if (drop_missing):
        df = df.dropna()  # drop all rows that have missing values in them


    X = np.vstack(df['Image'].values) / 255.  # scale pixel values to [0, 1]
    X = X.astype(np.float32)

    if not test:  # only FTRAIN has any target columns
        y = df[df.columns[:-1]].values
        y = (y - 48) / 48  # scale target coordinates to [-1, 1]
        X, y = shuffle(X, y, random_state=65539)  # shuffle train data using seed for consistency
        y = y.astype(np.float32)
    else:
        y = None

    return X, y

#For loading 2d data
def load2d(test=False, cols=None):
    X, y = load(test=test)
    X = X.reshape(-1, 1, 96, 96)
    return X, y
```

#####Loss Function

The goal of the network as it trains is to mimize the mean squared error loss function  $$ \operatorname{MSE}=\frac{1}{n}\sum_{i=1}^n(\hat{Y_i} - Y_i)^2.$$ where $\hat{Y_i}$ is the predicted value for sample $i$ and $Y_i$ is the actual value for sample $i$, and $n$ is the number of samples.

This is the one used by default in Lasagne NeuralNet. Kaggle evaluates submissions using the RMSE (Root mean squared error) however, as the RMSE punishes larger errors.



#####Activation Function

For the hidden layer I use a sigmoid activation function, which is defined as:  $f(t) = \dfrac{1}{1+e^{-t}}$  - you can find the implementation of it [here](https://github.com/Theano/Theano/blob/master/theano/tensor/nnet/sigm.py)

By default, the lasagne NeuralNet will use the rectifier activation function if none is provided. The rectifier activation function is defined as $f(x) = max(0,x)$ where x is the input to the neuron. This appears to be the a standard activation function lately in deep learning (According to discussions with Daniel Nouri and others on the lasagne github repo issues/commits comments) but I found that it didn't perform as well as the sigmoid activation function with the network I created. Here is the pseudocode for it, however:


    # rectify
    def rectify(x):
        # The following is faster than T.maximum(0, x),
        # and it works with nonsymbolic inputs as well.
        # Thanks to @SnipyHollow for pointing this out. Also see:
        # http://github.com/benanne/Lasagne/pull/163#issuecomment-81765117
        return 0.5 * (x + abs(x))


#####Number of hidden units


The number of hidden units is exactly what it sounds like: It is the number of units, aka neurons, that comprise the hidden layer. There have been I used GridSearchCV from sklearn in order to pick a number of hidden units for the network. How this works is, you pass in your estimator (The neural network) as well as a parameter grid and number of cross validation folds, and then it exhaustively searches over this grid of paramters for the best choice. My parameter grid was hidden unit numbers [100, 150, 200, 250, 300] and I used 2-fold cross validation. It found that 100 hidden units was the best performing. 

[I searched for a method to determine the hidden unit size, and found rule of thumb examples but there is no clear straightforward calculation to figure it out](http://stats.stackexchange.com/questions/181/how-to-choose-the-number-of-hidden-layers-and-nodes-in-a-feedforward-neural-netw)

Daniel Nouri (Who is an expert in this field) used 100 for his example 1-hidden layer MLP network, so I figured searching for an optimum parameter close to that would be a good idea. I chose to look at the paramater choices [100,150,..,300] because it is not feasible to try out [100,101,..,299,300] as for every single paramater choice I need to retrain the neural network.... If I wanted to use gridsearch cross validation with paramters [100,101,...,299,300] I would need to train (300-100+1) networks and that would take days...

Anyway, 100 did seem to be a good choice as I scored fairly well (#35 on the kaggle board) with that model.


#####Update Method

For an epoch, after the forward-pass has happened and the predicted values have been generated, the MSE is found and the weights are upadated accordingly. To updated, backpropagation happens (Theory in slides).

[Backpropagation methods](http://page.mi.fu-berlin.de/rojas/neural/chapter/K8.pdf)

I am using the [Nesterov's momentum - accelerated gradient descent algorithm](http://www.cs.toronto.edu/~fritz/absps/momentum.pdf) since it is a well performing optimization algorithm to improve the rate of learning.

######Momentum

The momentum coefficient is the fraction of the previous weight update to add to the current one. The values are between [0,1], and a higher momentum parameter can help increase the speed of convergence, but too high can risk overshooting the minimum. A small momentum parameter may make the network converge to a local minima as well as slow down the training. I've used the coefficient that Daniel Nouri uses, and later (For the next model) I will introduce the idea of slowly increasing the momentum while decreasing the learning rate instead of using a static coefficient for both.

#####Learning rate

This is essentially a coefficient determining how big of a step it takes in the new direction. If you use too high of a value it will converge to a local minimum too quickly and that will be that, I have used 0.01 as that is the paramter that Daniel Nouri used - and later, I will introduce the concept of adjusting the value as the network trains, decreasing it - since you want a higher learning rate at the start to get close to convergence and then a lower rate at the end to keep training it!

#####Output Nonlinearity

This is a misnomer. Lasagne NeuralNet has a nonlinearity paramter - this is the activation function of the output layer - however - it is a linear combination of the activations in the hidden layer, not a nonlinear function.

#####Early Stopping

The idea behind early stopping is that after you've been trainig your neural network for a while and the validation loss has not improved, that is a sign that it is not going to generalize any further, and you should stop early! How it works is you pass in a patience value, and after each epoch, it looks at the current validation loss, and if the current validation loss was worse than the (patience) number of epochs ago, it updates the weights to the weights of the best epoch (from the patience number of epochs ago) and stops training. This prevents overfitting as well as cuts training time down since the arbitrary number of epochs I am training to may not always be reached.



#####Putting it all together

here is the code that is used to create the network, perform gridsearchCV in order to choose the number of hidden units, and then finally, refit on the number of hidden units with the best score, and serialize to a binary file to be loaded later.

```python
#Neural Network # 1  - A MLP with 1 hidden layer
net1 = NeuralNet(
    layers=[  # three layers: one hidden layer
        ('input', layers.InputLayer),
        ('hidden', layers.DenseLayer),
        ('output', layers.DenseLayer),
        ],
    # layer parameters:
    input_shape=(None, 9216),  # 96x96 input pixels per batch
    hidden_num_units=100,  # number of units in hidden layer
    hidden_nonlinearity=sigmoid,
    output_nonlinearity=None,  # output layer uses identity function
    output_num_units=30,  # 30 target values
    # optimization params
    update=nesterov_momentum,
    update_learning_rate=0.01,
    update_momentum=0.9,
    regression=True, 
    max_epochs=1000,
    on_epoch_finished=[
        EarlyStopping(patience=50)
        ],
    verbose=1 #It will output the summary of each epoch if set to 1 - takes up a lot of memory in firefox if you keep it like this!
    )


#paramter grid for gridsearch cv
param_grid = {
'more_params': [{'hidden_num_units': 100}, {'hidden_num_units':150}, {'hidden_num_units': 200},
{'hidden_num_units': 250}, {'hidden_num_units': 300}]
}
 
#find net with best params , and then refit with all data on best net
gs = GridSearchCV(net1, param_grid, cv=2, refit=True, verbose=4)

#Load data
X,y=load()
#Fit CV model
gs.fit(X,y)

#serialize model
with open('net1_gridsearch.pickle', 'wb') as f:
    # we serialize the gridsearch model, as it also has the fitted neural network with best hidden unit number:
    pickle.dump(gs, f, -1)
    #if you just want to save the best estimator - ie - the neural network itself - without the gridsearch info:
    #pickle.dump(gs.best_estimator_, f, -1)
    
    
```

The following is the sample Early Stopping class, courtesy of Daniel Nouri: If you want to load the pickled neural net, you need to run this cell, as the pickled net includes an instantiated earlystopping object inside of it and it wont load of the class is not defined in the program.

```python
class EarlyStopping(object):
    def __init__(self, patience=100):
        self.patience = patience
        self.best_valid = np.inf
        self.best_valid_epoch = 0
        self.best_weights = None

    def __call__(self, nn, train_history):
        current_valid = train_history[-1]['valid_loss']
        current_epoch = train_history[-1]['epoch']
        if current_valid < self.best_valid:
            self.best_valid = current_valid
            self.best_valid_epoch = current_epoch
            self.best_weights = [w.get_value() for w in nn.get_all_params()]
        elif self.best_valid_epoch + self.patience < current_epoch:
            print("Early stopping.")
            print("Best valid loss was {:.6f} at epoch {}.".format(
                self.best_valid, self.best_valid_epoch))
            nn.load_weights_from(self.best_weights)
            raise StopIteration()


##These are for later...
#For shared learning/update rates
def float32(k):
    return np.cast['float32'](k)

            
#http://danielnouri.org/notes/2014/12/17/using-convolutional-neural-nets-to-detect-facial-keypoints-tutorial/#changing-learning-rate-and-momentum-over-time
class AdjustVariable(object):
    def __init__(self, name, start=0.03, stop=0.001):
        self.name = name
        self.start, self.stop = start, stop
        self.ls = None

    def __call__(self, nn, train_history):
        if self.ls is None:
            self.ls = np.linspace(self.start, self.stop, nn.max_epochs)

        epoch = train_history[-1]['epoch']
        new_value = float32(self.ls[epoch - 1])
        getattr(nn, self.name).set_value(new_value)
```

#### Learning Curve

Here's a function for plotting the learning curve for our models..

```python
#Plot learning curve using specified loss function (RMSE or MSE)
def plot_acc(net, label, loss='RMSE'):
    if loss == 'RMSE':
        train_loss = np.sqrt(np.array([i["train_loss"] for i in net.train_history_]))*48.
        valid_loss = np.sqrt(np.array([i["valid_loss"] for i in net.train_history_]))*48.
        plt.ylim(0,5)

    elif loss == 'MSE':
        train_loss = np.array([i["train_loss"] for i in net.train_history_])
        valid_loss = np.array([i["valid_loss"] for i in net.train_history_])
        plt.yscale("log")

    plt.plot(train_loss, linewidth=2, label="train loss")
    plt.plot(valid_loss, linewidth=2, label="valid loss")
    plt.grid()
    plt.legend()
    plt.xlabel("epoch")
    plt.ylabel(loss)

    plt.suptitle(loss+ " learning curve for "+label)
    plt.show()    

plot_acc(net1, "net 1:", 'MSE')
plot_acc(net1, "net 1:", 'RMSE')

```

This is simply plotting the loss over each epoch, for the MSE as well as RMSE (Kaggle's evaluation)

```python
def plot_accs(nets, labels, loss='RMSE'):
    fig = plt.figure()
    for x in range(6):
        
        if loss == 'RMSE':
            train_loss = np.sqrt(np.array([i["train_loss"] for i in nets[x].train_history_]))*48.
            valid_loss = np.sqrt(np.array([i["valid_loss"] for i in nets[x].train_history_]))*48.
            plt.ylim(0, 5)
    
        elif loss == 'MSE':
            train_loss = np.array([i["train_loss"] for i in nets[x].train_history_])
            valid_loss = np.array([i["valid_loss"] for i in nets[x].train_history_])
            plt.yscale("log")
            plt.ylim((1e-3*0.5), 1e-1*0.2)
        plt.plot(train_loss, linewidth=2, label="train loss for: "+labels[x], ls='--')
        plt.plot(valid_loss, linewidth=2, label="valid loss for: "+labels[x])
    fig.set_size_inches(16,12)
    plt.grid()
    plt.xlabel("epoch - is different for each specialized network due to early stopping")
    plt.ylabel(loss)
    plt.suptitle(loss+ " learning curve for specialized networks")
    plt.legend(loc='best')
    plt.show()    

plot_accs([x[1] for x in net2], labels=net2_labels)
```
